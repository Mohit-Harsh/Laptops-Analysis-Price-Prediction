```python
browser.get(acer)

# execute java script
browser.execute_script("return document.getElementsByTagName('html')[0].innerHTML")

# wait page to load
sleep(1)
```

```python
soup = BeautifulSoup(browser.page_source,'html.parser')
```

```python
navs = [x['href'] for x in soup.find('nav',{'class':'yFHi8N'}).find_all('a')[:-1]]
navs
```

```python
laptops = []
```

```python
for nav in navs:
    
    URL = "https://www.flipkart.com"+nav

    browser.get(URL)
    browser.execute_script("return document.getElementsByTagName('html')[0].innerHTML")
    sleep(1)    

    s = BeautifulSoup(browser.page_source,'html.parser')
    
    laptops += list([x.find('a')['href'] for x in s.find_all('div',{"class":"_1AtVbE col-12-12"})[2:-2]])
```

```python
l = len(laptops)
brand,processor_brand,processor_name,ssd,ram,ram_type,clock,gpu,gpu_memory = [None]*l,[None]*l,[None]*l,[None]*l,[None]*l,[None]*l,[None]*l,[None]*l,[None]*l
price = [None]*l
c=0
print(l)
```

```python
for lap in laptops:
    
    URL = "https://www.flipkart.com"+lap

    browser.get(URL)
    browser.execute_script("return document.getElementsByTagName('html')[0].innerHTML")
    sleep(1)

    sp = BeautifulSoup(browser.page_source,'html.parser')

    features = sp.find_all('tr',{'class':'_1s_Smc row'})
    
    if features:
        
        brand[c] = sp.find('h1',{'class':'yhB1nd'}).text.split()[0].strip()
        price[c] = sp.find('div',{'class':'_30jeq3 _16Jk6d'}).text.strip()

        for f in features:
            
            t = f.find_all('td')
            
            if t[0].text == "Processor Brand":
                
                processor_brand[c] = t[1].text
                
            elif t[0].text == "Processor Name":
                
                processor_name[c] = t[1].text
                
            elif t[0].text == "SSD Capacity":
                
                ssd[c] = t[1].text
                
            elif t[0].text == "RAM":
                
                ram[c] = t[1].text
                
            elif t[0].text == "RAM Type":
                
                ram_type[c] = t[1].text
                
            elif t[0].text == "Clock Speed":
                
                clock[c] = t[1].text
                
            elif t[0].text == "Graphic Processor":
                
                gpu[c] = t[1].text
                
            elif t[0].text == "Dedicated Graphic Memory Capacity":
                
                gpu_memory[c] = t[1].text
                
        c = c + 1
        
    else:
        
        continue
```

```python
data = pd.DataFrame()
```

```python
data['brand'] = brand
data['processor_brand'] = processor_brand
data['processor_name'] = processor_name
data['ram'] = ram
data['ram_type'] = ram_type
data['ssd'] = ssd
data['clock'] = clock
data['gpu'] = gpu
data['gpu_memory'] = gpu_memory
data['price'] = price
```

```python
data['gpu_memory'].fillna(0,inplace=True)
```

```python
data.dropna(inplace=True)
data.info()
```

```python
data = data.reset_index().drop(['index'],axis=1)
```
